# Creating Thumbnails of WSIs(Whole Slide Images)

### 1. Import Packages and Functions

```python
from openslide import open_slide
import openslide
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

### 2. Properties and Thumbnail

```python
# Read image using open_slide
slide=open_slide("/home/vipul/Downloads/gdc_download_20220821_062703.958218/75bba1b4-3fe5-4657-a5a9-0edc643b44fd/TCGA-18-3409-01Z-00-DX1.3E32CC78-B066-4F85-A781-EDAF39DE0704.svs")

# checking the properties of the image
slide_props=slide.properties
print(list(slide_props))
print()

#getting the vendor of image
print("Vendor is ",slide_props['openslide.vendor'])

# Dimensions of the slide
slide_dims=slide.dimensions
print(slide_dims)

# Get a thumbnail of the image and visualize
slide_thumb=slide.get_thumbnail(size=(200,200))
print(type(slide_thumb))
#slide_thumb.show()
#im1 =slide_thumb.save("thumbnail.jpg")
print("Done")


```

### 3. Display the image

```python
#convert thumbnail to numpy array
slide_thumb_np=np.array(slide_thumb)
plt.figure(figsize=(8,8))
plt.imshow(slide_thumb_np)
```
